### **Lab Task:** Using sum() for Power and Resistance
### **Author:** Hamdan
### **Purpose:** Demonstrate sum() for total power and series resistance

#### Step 1: Total power consumed

```python
power_readings = [10, 20, 15, 5]   # watts
total_power = sum(power_readings)
print("Total Power Consumed:", total_power, "W")
```

#### Step 2: Equivalent resistance in series

```python
resistances = [4, 6, 10, 5]   # ohms
equivalent_resistance = sum(resistances)
print("Equivalent Series Resistance:", equivalent_resistance, "ohms")
```
